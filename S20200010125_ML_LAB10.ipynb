```python
x=[]
y=[]
for i in range(60,301):
  x.append(i*(math.pi/180))
a,s=0,0.15
for i in range(len(x)):
  K = np.random.normal(a,s)
  Z = math.sin(x[i])+K
  y.append(Z)
a0,a1=0,0
for j in range(100):
  j0,j1 = 0,0
  theta=[]
  for i in range(len(y)):
    j0+=(-1*(y[i]-(a0+a1*x[i])))*2*(-0.0001)
    j1+=(-x[i]*(y[i]-(a0+a1*x[i])))*2*(-0.0001)
  theta.append(a0+j0)
  theta.append(a1+j1)
  a0=theta[0]
  a1=theta[1]
x1=np.array(x)
y1 = theta[0] + theta[1]*x1
plt.plot(x1,y1)
plt.scatter(x,y,color='red')
sse=0
for i in range(len(y)):
  sse+=((y1[i]-y[i])**2)
print("SSE :",sse)
print("Co-Efficients : " ,theta[0],theta[1])
plt.show()
```
