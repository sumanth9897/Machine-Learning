```python
from tensorflow.keras.datasets import mnist
import numpy as np
import matplotlib.pyplot as plt
tree, types = mnist.load_data()
```

```python
x_tree = tree[0]
y_tree = tree[1]
x_types = types[0]
y_types = types[1]
```

```python
i=4
plt.imshow(x_tree[i])
```

```python
import tensorflow as tf
from tensorflow.keras.layers import Dense,Flatten
from tensorflow.keras.models import Sequential
```

```python
copy = Sequential([
        Flatten(input_shape=(28,28)),
        Dense(64, activation='relu'),
        Dense(32, activation='relu'),
        Dense(10, activation='softmax'),
    ])

copy.compile(optimizer = 'adam',
              loss='SparseCategoricalCrossentropy', metrics=['accuracy'])
history = copy.fit(x_tree, y_tree, validation_split=0.2, epochs=20, verbose=1)
copy.evaluate(x_types, y_types)
```

```python
perfect = [0.001,0.003,0.006,0.008,0.01]
```

```python
def plan(history):
    plt.figure(facecolor='grey')
    l1 = plt.plot(history.history['loss'])
    l2 = plt.plot(history.history['val_loss'])
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    labels = ["Loss", "Validation Loss"]
    plt.legend([l1, l2], labels=labels,
           loc="upper right")
```

```python
def MD_Ke():
        copy = Sequential([
        Flatten(input_shape=(28,28)),
        Dense(64, activation='relu'),
        Dense(32, activation='relu'),
        Dense(10, activation='softmax'),
    ])
        return copy
```

```python
def attach(lr):
    copy = MD_Ke()
    copy.compile(optimizer = tf.keras.optimizers.Adam(learning_rate=lr),
                  loss='SparseCategoricalCrossentropy', metrics=['accuracy'])
    history = copy.fit(x_tree, y_tree, validation_split=0.2, epochs=20, verbose=1)
    copy.evaluate(x_types, y_types)
    return history
```

```python
history = []
for i in range(len(perfect)):
    history.append(attach(perfect[i]))
```

```python
for i in range(len(history)):
    plan(history[i])
```
